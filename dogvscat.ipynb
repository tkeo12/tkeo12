```python
import tensorflow as tf
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    'dataset',
    image_size=(64,64),
    batch_size=64,               #이미지 2만장 한번에 넣지 않고 batch 숫자만큼 넣음
    subset='training',
    validation_split=0.2,
    seed=1234
) # 이미지 숫자화 함수 , 폴더 내 이미지들을 바로 dataset 만들어줌

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    'dataset',
    image_size=(64,64),
    batch_size=64,               #이미지 2만장 한번에 넣지 않고 batch 숫자만큼 넣음
    subset='validation',
    validation_split=0.2,
    seed=1234
) # 이미지 숫자화 함수 , 폴더 내 이미지들을 바로 dataset 만들어줌

print(train_ds)

def 전처리함수(i,정답):
    i = tf.cast( i/255.0, tf.float32)
    return i, 정답

train_ds=train_ds.map(전처리함수)
val_ds=val_ds.map(전처리함수)
```

```python
# 새로운 경로를 설정합니다. 이 폴더는 미리 생성되어 있어야 합니다.
new_path = "C:\\Users\\tkeo1\\OneDrive\\바탕 화면\\coding\\Python\\data\\validation"

# 파일 경로와 레이블을 가져옵니다.
file_paths = val_ds.file_paths

for file_path in file_paths:
    # 파일을 새 경로로 이동합니다. 이것은 잘라내기 및 붙여넣기 작업에 해당합니다.
    shutil.move(file_path, os.path.join(new_path, os.path.basename(file_path)))
```

```python
model=tf.keras.Sequential([
    
    tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal',input_shape=(128,128,3)),
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.1),
    tf.keras.layers.experimental.preprocessing.RandomZoom(0.1),
    
    tf.keras.layers.Conv2D(32,(3,3),padding="same",activation='relu'), #이미지데이터는 0~255 숫자 , 음수 내보내지 않기 위해 // input_shape 컬러사진일때는 28,28,3 (RGB 때문)
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(64,(3,3),padding="same",activation='relu'), #이미지데이터는 0~255 숫자 , 음수 내보내지 않기 위해 // input_shape 컬러사진일때는 28,28,3 (RGB 때문)
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Conv2D(128,(3,3),padding="same",activation='relu'), #이미지데이터는 0~255 숫자 , 음수 내보내지 않기 위해 // input_shape 컬러사진일때는 28,28,3 (RGB 때문)
    tf.keras.layers.MaxPooling2D((2,2)),
    # tf.keras.layers.Dense(128,input_shape=(28,28),activation="relu"),
    tf.keras.layers.Flatten(), # 2d data -> 1d data
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation="sigmoid"), # convol 2D > pool layer > flatten > dense
])  
#sigmoid = binary prediction (finally number of node = 1)
#softmax = category prediction (finally number of node > 1)
model.summary()

model.compile(loss="binary_crossentropy", optimizer="adam",metrics=['accuracy'])
# sparse_categorical_crossentropy = category prediction 정수로 인코딩 되어있을때 (0,1,2,...)
# categorical_crossentropy = category prediction 는 Y데이터가 원핫인코딩 되어있을때 (0 , 1)

model.fit(train_ds,validation_data=val_ds,epochs=20) # train_ds 형태는 ([이미지],[정답])
```
