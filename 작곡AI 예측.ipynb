```python
import tensorflow as tf

Pmodel=tf.keras.models.load_model('./model')

text = open('C:/Users/tkeo1/OneDrive/바탕 화면/coding/Python/data/pianoabc.txt','r').read()
# print(text)

text_uq = list(set(text)) # 리스트랑 비슷한데 중복자료 허용 x
text_uq. sort()
# print(text_uq)

# utilities
text_to_num = {}
num_to_text={}

for i, data in enumerate(text_uq):
    text_to_num[data]=i
    num_to_text[i]=data

# print(text_to_num)
print(num_to_text[3])

text_nu=[] # 원본데이터를 숫자로 바꾸기

for i in text:
    text_nu.append(text_to_num[i])

# print(text_nu)
```

```python
첫입력값=text_nu[117 : 117 + 25]
첫입력값=tf.one_hot(첫입력값,31)
첫입력값=tf.expand_dims(첫입력값, axis=0) # 2차원 -> 3차원
print(첫입력값)

```

```python
예측값=Pmodel.predict(첫입력값)
import numpy as np
np.argmax(예측값[0])
print(예측값)
print(text_nu[117 : 117 + 25])
```

```python
# 1. 첫입력값 만들기
# 2. predict로 다음문자 예측

# 3. 예측한 다음문자 [] 저장하기
# 4. 첫입력값 앞에 자르기
# 5. 예측한 다음문자를 뒤에 넣기
# 6. 원핫 인코딩, expand dims
# 6. 그 새로운 입력값을 predict

```

```python
music=[]
import numpy as np
for i in range(200):
    예측값=Pmodel.predict(첫입력값)
    예측값=np.argmax(예측값[0])
    # 로운예측값=np.random.choice(text_uq,1,p=예측값[0])
    #print(예측값)
    
    music.append(예측값)
    다음입력값=첫입력값.numpy()[0][1:]
    #print(다음입력값)
    
    one_hot_num=tf.one_hot(예측값,31)
    #print('원핫한거',one_hot_num)
    
    첫입력값=np.vstack([다음입력값 , one_hot_num.numpy()])
    첫입력값=tf.expand_dims(첫입력값,axis=0)

print(music)
```

```python
music_text=[]
for i in music:
    music_text.append(num_to_text[i])


print(''.join(music_text))
```
